### Importación de datos



````python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

df_tienda1 = pd.read_csv(url)
df_tienda2 = pd.read_csv(url2)
df_tienda3 = pd.read_csv(url3)
df_tienda4 = pd.read_csv(url4)

df_tienda1.head()
````

#1. Análisis de facturación



````python
ingresos_tienda1 = (df_tienda1['Precio'] + df_tienda1['Costo de envío']).sum()
ingresos_tienda2 = (df_tienda2['Precio'] + df_tienda2['Costo de envío']).sum()
ingresos_tienda3 = (df_tienda3['Precio'] + df_tienda3['Costo de envío']).sum()
ingresos_tienda4 = (df_tienda4['Precio'] + df_tienda4['Costo de envío']).sum()

print(f"Ingresos Tienda 1: ${ingresos_tienda1:,.2f}")
print(f"Ingresos Tienda 2: ${ingresos_tienda2:,.2f}")
print(f"Ingresos Tienda 3: ${ingresos_tienda3:,.2f}")
print(f"Ingresos Tienda 4: ${ingresos_tienda4:,.2f}")
````

# 2. Ventas por categoría

````python
ingresos_categoria_t4 = df_tienda4.groupby('Categoría del Producto')['Precio'].sum().sort_values(ascending=False)

print("--- Ingresos por Categoría - Tienda 4 ---")
print(ingresos_categoria_t4)
````

````python
ingresos_categoria_t1 = df_tienda1.groupby('Categoría del Producto')['Precio'].sum().sort_values(ascending=False)

print("\n--- Ingresos por Categoría - Tienda 1 ---")
print(ingresos_categoria_t1)
````

````python
#precio promedio por categoría para cada tienda
precio_promedio_t1 = df_tienda1.groupby('Categoría del Producto')['Precio'].mean()
precio_promedio_t4 = df_tienda4.groupby('Categoría del Producto')['Precio'].mean()

#DataFrame para comparar
comparacion_precio_promedio_df = pd.DataFrame({
    'Tienda 1': precio_promedio_t1,
    'Tienda 4': precio_promedio_t4
})

print(comparacion_precio_promedio_df)
````

# 3. Calificación promedio de la tienda

## Informe de Análisis y Recomendación para Alura Store

**Introducción**: Propósito del Análisis

El objetivo de este análisis es realizar una evaluación comparativa del rendimiento de las cuatro sucursales de la cadena Alura Store. A través del estudio de métricas clave de ventas, costos y satisfacción del cliente, se busca identificar la tienda con el rendimiento más bajo para proporcionar una recomendación informada al Sr. Juan sobre cuál sucursal sería la más indicada para vender.

**Desarrollo**: Análisis de los Datos
Para llegar a una conclusión, empecé analizando los indicadores más importantes y luego profundicé en los detalles para entender el porqué de los resultados.

1. **Ingresos Totales**: El Indicador Principal
Lo primero que analicé fueron los ingresos totales de cada tienda, ya que es la métrica más directa del éxito financiero. Los resultados muestran una clara diferencia en el rendimiento:

```
Tienda 1: R$ 1,212,258,300.00
Tienda 2: R$ 1,175,828,600.00
Tienda 3: R$ 1,156,536,200.00
Tienda 4: R$ 1,093,693,100.00
```

*Como se puede ver, la Tienda 4 es la que menos ingresos genera. Para cuantificar esta diferencia, calculé que la Tienda 4 vende un -9.78% menos que la Tienda 1, la sucursal con mejor rendimiento.*

grafico1.png

2. **Calificaciones y Costos de Envío**: Buscando el Contexto
Luego, investigué si la satisfacción del cliente podría explicar esta diferencia. Los resultados de las calificaciones promedio fueron sorprendentes, ya que son muy similares en todas las tiendas, rondando las 4.0 estrellas.

```
Calificación Promedio Tienda 1: 3.98 estrellas
Calificación Promedio Tienda 4: 4.00 estrellas
```
*Esto me dice que el problema de la Tienda 4 probablemente no es una mala reputación o un mal servicio al cliente.*

grafico2.png

*Por otro lado, al analizar los costos de envío, noté que la Tienda 4 tiene el envío promedio más bajo de todas. Mi hipótesis era que esto podría atraer a clientes locales, pero parece que esta ventaja no ha sido suficiente para impulsar los ingresos al nivel de las otras tiendas.*

3. **Ventas por Categoría**: ¿Dónde está el Problema Real?
Para entender la causa raíz del bajo rendimiento de la Tienda 4, analicé de dónde provienen sus ingresos, comparándolos con los de la Tienda 1. Aquí encontré la clave del análisis.

Si bien la Tienda 4 es competitiva en categorías de valor moderado como "Muebles", **muestra una debilidad crítica en las categorías que más dinero generan**: "Electrónicos" y "Electrodomésticos". La diferencia de ingresos en estas dos categorías entre la Tienda 1 y la Tienda 4 es de decenas de millones, lo que explica la mayor parte del déficit total de la tienda.

##Conclusión y Recomendación Final
Con base en los análisis realizados, mi recomendación para el Sr. Juan es vender la Tienda 4.

Esta decisión se fundamenta en los siguientes puntos clave:

*Menor Rendimiento Financiero*: Es, indiscutiblemente, la tienda que menos ingresos genera, con casi un 10% menos que la tienda líder.
Debilidad en Categorías Estratégicas: Su principal problema es la incapacidad de competir en las categorías más lucrativas (Electrónicos y Electrodomésticos), que son el motor financiero de las otras sucursales.
Ventajas no Capitalizadas: A pesar de tener el costo de envío más bajo, esta ventaja no se traduce en un rendimiento de ventas superior que compense sus debilidades en otras áreas.
Ausencia de Otros Diferenciadores: No posee una calificación de clientes notablemente superior que justifique mantenerla a pesar de su bajo rendimiento financiero.

*Por lo tanto, vender la Tienda 4 parece ser la decisión de negocio más lógica para permitir al Sr. Juan enfocarse en sus tiendas más rentables e iniciar su nuevo emprendimiento.*
